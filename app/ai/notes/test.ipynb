```python
from paddleocr import PaddleOCR
from PIL import Image

# 이미지 경로 설정
img_path = "../../test/images/realtest.jpg"

# 한국어로 PaddleOCR 초기화
ocr = PaddleOCR(use_angle_cls=True, lang='korean')

# 이미지에서 텍스트 인식
result = ocr.ocr(img_path, cls=True)

# 이미지 열기
image = Image.open(img_path).convert('RGB')

for line in result:
    for word in line:
        print(word[1])
```
